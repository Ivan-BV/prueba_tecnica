# Creación de la bbdd

Tablas: Peliculas, Artistas, Reparto, Calificaciones

`Peliculas` tiene relacion con `Reparto` y `Calificaciones`

`Reparto` tiene relación con `Artistas`

```python
try:
    connection = psycopg2.connect(
        dbname="postgres",
        user="postgres",
        password="admin",
        host="localhost",
    )
    connection.autocommit = True
    nombre_bbdd = "prueba_tecnica"

    # Crear un cursor para ejecutar comandos SQL
    cursor = connection.cursor()

    cursor.execute("SELECT 1 FROM pg_database WHERE datname = %s", (nombre_bbdd,))
    
    # Almacenamos en una variable el resultado del fetchone. Si existe tendrá una fila sino será None
    bbdd_existe = cursor.fetchone()
    
    # Si bbdd_existe es None, creamos la base de datos
    if not bbdd_existe:
        cursor.execute("CREATE DATABASE prueba_tecnica;")
        print(f"Base de datos {nombre_bbdd} creada con éxito")
    else:
        print(f"La base de datos ya existe")

    connection = psycopg2.connect(
        dbname=nombre_bbdd,
        user="postgres",
        password="admin",
        host="localhost",
    )

    # Definir la estructura de las tablas
    query_create_peliculas = """
    CREATE TABLE IF NOT EXISTS peliculas (
        id_peli INT PRIMARY KEY,
        titulo VARCHAR(300),
        tipo VARCHAR(300),
        anio NUMERIC,
        mes NUMERIC,
        generos VARCHAR(300)
    )"""
    cursor.execute(query_create_peliculas)

    query_create_artistas = """
    CREATE TABLE IF NOT EXISTS artistas (
        id_artista SERIAL PRIMARY KEY,
        nombre VARCHAR(300),
        anio NUMERIC,
        conocido_por VARCHAR(300),
        que_hace VARCHAR(300),
        premios VARCHAR(300)
    )"""
    cursor.execute(query_create_artistas)

    query_create_reparto = """
    CREATE TABLE IF NOT EXISTS reparto (
        id_peli INT PRIMARY KEY REFERENCES peliculas(id_peli),
        id_artista INT PRIMARY KEY REFERENCES artistas(id_artista)
    )"""
    cursor.execute(query_create_reparto)

    query_create_calificaciones = """
    CREATE TABLE IF NOT EXISTS calificaciones (
        id_peli INT PRIMARY KEY REFERENCES peliculas(id_peli),
        titulo VARCHAR(300),
        puntuacion VARCHAR(300),
        director VARCHAR(300),
        guionistas VARCHAR(300),
        argumento VARCHAR(600),
        duracion VARCHAR(300)
    )"""
    cursor.execute(query_create_calificaciones)
    
except Exception as e:
    connection.rollback()
    print("Error al conectar a la base de datos:", e)
    raise
```
